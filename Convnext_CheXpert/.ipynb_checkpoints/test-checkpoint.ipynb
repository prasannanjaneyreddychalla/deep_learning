```python
import pandas as pd
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms
from PIL import Image

# Define the path to your dataset and CSV file
DATA_DIR = "/scratch/smanika3/chexpert/full_uncompressed"
TRAIN_CSV = f"{DATA_DIR}/train_cheXbert.csv"

# Load the CSV file and handle uncertain labels
# A common strategy is to treat 'uncertain' (-1) as positive (1) for some pathologies
# while others are treated as negative (0). Here we just treat them as positive.
df = pd.read_csv(TRAIN_CSV)
df = df.fillna(0)  # Fill NaNs with 0
df.replace(to_replace=-1, value=1, inplace=True) # Replace -1 with 1

# Extract the labels for the 14 pathologies
pathologies = df.columns[5:].tolist()
```

```python
class ChexpertDataset(Dataset):
    def __init__(self, dataframe, data_dir, transform=None):
        self.dataframe = dataframe
        self.data_dir = data_dir
        self.transform = transform
        self.pathologies = dataframe.columns[5:].tolist()
        
    def __len__(self):
        return len(self.dataframe)

    def __getitem__(self, idx):
        # The path in the CSV is relative, so we join it with the data directory
        img_path = f"{self.data_dir}/{self.dataframe.iloc[idx]['Path']}"
        image = Image.open(img_path).convert('RGB')
        
        labels = self.dataframe.iloc[idx][self.pathologies].values.astype(float)
        labels = torch.tensor(labels, dtype=torch.float32)

        if self.transform:
            image = self.transform(image)
        
        return image, labels

# Define the image transformations
# ConvNeXt expects 224x224 input by default
image_size = 224
data_transforms = transforms.Compose([
    transforms.Resize((image_size, image_size)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
])
```

```python
import timm
from torch import nn, optim

# Load a pre-trained ConvNeXt model from timm
model = timm.create_model('convnext_base', pretrained=True)

# Adapt the classifier head for the CheXpert task (14 output classes)
num_pathologies = len(pathologies)
model.head.fc = nn.Linear(model.head.fc.in_features, num_pathologies)

# Use Binary Cross Entropy Loss for the multi-label problem
criterion = nn.BCEWithLogitsLoss()

# Use the AdamW optimizer, as it is standard for training ConvNeXt
optimizer = optim.AdamW(model.parameters(), lr=1e-4)

# Set up data loaders
batch_size = 32
# Create the dataset instance with the modified path
dataset = ChexpertDataset(dataframe=df, data_dir="", transform=data_transforms)
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Select the device for training (GPU if available)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
```

```python
num_epochs = 1  # For a quick example, increase for real training

for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for i, (images, labels) in enumerate(dataloader):
        images, labels = images.to(device), labels.to(device)

        # Zero the parameter gradients
        optimizer.zero_grad()

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward pass and optimize
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        if (i + 1) % 100 == 0:
            print(f"Epoch [{epoch+1}/{num_epochs}], Step [{i+1}/{len(dataloader)}], Loss: {running_loss/100:.4f}")
            running_loss = 0.0

    print(f"Finished training for epoch {epoch+1}")
```

```python
# Create the validation dataset and dataloader
valid_df = pd.read_csv(f"{DATA_DIR}/valid.csv")
valid_df.replace(to_replace=-1, value=1, inplace=True)
valid_df = valid_df.fillna(0)
valid_dataset = ChexpertDataset(dataframe=valid_df, data_dir="chexpert-dataset", transform=data_transforms)
valid_dataloader = DataLoader(valid_dataset, batch_size=batch_size)

model.eval()
validation_loss = 0.0
with torch.no_grad():
    for images, labels in valid_dataloader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        loss = criterion(outputs, labels)
        validation_loss += loss.item()

print(f"Validation Loss: {validation_loss / len(valid_dataloader):.4f}")
```
